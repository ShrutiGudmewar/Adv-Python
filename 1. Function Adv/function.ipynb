```python
def div(x: float| int, y: float|int)-> int|float:
    """The above syntax define the type of args this fun will accept and 
    the type of return value. Which is either float or int """
    assert isinstance(x,(int,float)),"Expected x to be integer or float"
    assert isinstance(y,(int,float)),"Expexted y to be int or float"
    """above sytax check wheather the x and y are float or int: if the are not float or int it will show the give 
    string which is Expected x or y to be int or float"""
    z=x/y
    return z
```

```python
div(10, 2)
```

```python
div(20,2)
```

```python
#Arguments
div(23,7)
```
